```python
## Task 2: Mark the success/failed launches for each site on the map

# This code assumes 'site_map' from Task 1 and 'spacex_df' are already created and available.

# 1. Create a MarkerCluster object
marker_cluster = MarkerCluster()

# 2. Create a new 'marker_color' column in the spacex_df dataframe
# Use a lambda function to assign 'green' for class 1 (success) and 'red' for class 0 (failure)
spacex_df['marker_color'] = spacex_df['class'].apply(lambda c: 'green' if c == 1 else 'red')

# 3. Add the MarkerCluster to the site_map
site_map.add_child(marker_cluster)

# 4. Iterate through each row in the spacex_df dataframe
for index, record in spacex_df.iterrows():
    # Get the coordinates and other info for each launch
    launch_site_name = record['Launch Site']
    coordinate = [record['Lat'], record['Long']]
    outcome_class = record['class']
    marker_color = record['marker_color']
    
    # Create a Marker object for each record
    marker = folium.Marker(
        location=coordinate,
        # Create a popup with useful information
        popup=f"Site: {launch_site_name}<br>Outcome: {'Success' if outcome_class==1 else 'Failure'}",
        # Customize the icon's color based on the outcome
        icon=folium.Icon(color='white', icon_color=marker_color)
    )
    
    # Add the marker to the marker_cluster, not the map directly
    marker_cluster.add_child(marker)

# Display the final map with marker clusters
site_map
```

Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
## TASK 3: Calculate the distances between a launch site to its proximities

# This assumes 'site_map' from Task 2 is available.

# 1. Add MousePosition to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)
site_map.add_child(mouse_position)
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
# 2. Define the coordinates for the launch site and the point of interest
# Get launch site coordinates from our launch_sites_df
launch_site_lat = 28.563197
launch_site_lon = -80.576820
launch_site_coords = [launch_site_lat, launch_site_lon]

# Manually find the closest coastline coordinates using the MousePosition tool on the map
coastline_lat = 28.56367
coastline_lon = -80.56794
coastline_coords = [coastline_lat, coastline_lon]


# 3. Calculate the distance between the two points
distance = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
print(f"Distance to coastline: {distance:.2f} KM")
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property
distance_marker = folium.Marker(
   location=coastline_coords,
   icon=DivIcon(
       icon_size=(20,20),
       icon_anchor=(0,0),
       html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
       )
   )
site_map.add_child(distance_marker)
```
